```python
def mean(arr):
  tot=0
  for i in range(len(arr)):
    tot+=arr[i]
  return tot/len(arr)
def sd(arr):
  tot=0
  x=mean(arr)
  for i in range(len(arr)):
    tot+=(arr[i]-x)**2
  result=(tot/(len(arr)-1))**0.5
  return round(result,2)
def new_array(arr):
  x=mean(arr)
  u=sd(arr)
  count=0
  for i in range(len(arr)):
    if arr[i]<(x-1.5*u) or arr[i]>(x+1.5*u):
      count+=1
  if count==0:
    return None
  result=np.array([0]*count)
  for i in range(len(arr)):
    if arr[i]<(x-1.5*u) or arr[i]>(x+1.5*u):
      result[-count]=arr[i]
      count-=1
  return result



sample=np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(f"The mean of the numbers is: {mean(sample)}")
print(f"The standard deviation is: {sd(sample)}")
print(f"New array: {new_array(sample)}")
```

```python
def compress_matrix(mat):
    #TO DO
    row,col=mat.shape
    arr=np.zeros((row//2,col//2),dtype=int)
    for i in range(0,row,2):
      for j in range(0,col,2):
        arr[i//2][j//2]=mat[i][j]+mat[i+1][j]+mat[i][j+1]+mat[i+1][j+1]
    return arr



matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

```python
#Run this cell
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

```python
class VehicleHashTable:
  def __init__(self, size):
    self.vehicleTable = [None]*size
    self.empty_slot = {}
    for i in range(size):
      self.empty_slot[i] = False

  def create_from_vehicle_info_array(self, arr):
    for i in arr:
      self.insert_vehicle(i)

  def print_vehicle_hashtable(self):
    idx = 0
    for i in self.vehicleTable:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(Brand: {head.brand}, Type: {head.vehicle_type}, Rent: {head.rent}, No. of Passengers: {head.passenger})', end = '---->')
        head = head.next
      print('None')
      print()
      idx += 1

  def find_empty_slot(self):
    for k,v in self.empty_slot.items():
      idx = k
      del self.empty_slot[k]
      return idx


  #Do it by yourself
  def __hash_function(self, brand):
      if brand=="":
        return 0
      val=ord(brand[0])
      index=val+self.__hash_function(brand[1:])
      index=index%len(self.vehicleTable)
      # Do not discard the next 3 lines
      if index in self.empty_slot:
        del self.empty_slot[index]
      return index


  #Do it by yourself
  def insert_vehicle(self, vehicle):
    #TO DO
    node=vehicleNodes(vehicle[0],vehicle[1],vehicle[2],vehicle[3])
    idx=self.__hash_function(node.brand)
    if self.vehicleTable[idx]==None:
      self.vehicleTable[idx]=node
    elif node.brand==self.vehicleTable[idx].brand:
      node.next=self.vehicleTable[idx]
      self.vehicleTable[idx]=node
    else:
      for i in range(len(self.vehicleTable)):
        if self.vehicleTable[i]==None:
          self.vehicleTable[i]=node
          break

    return self.vehicleTable
```
